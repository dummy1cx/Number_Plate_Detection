###### Test 1 .jpg image with unseen image

```python
import cv2
import matplotlib.pyplot
import ultralytics
```

```python
import matplotlib.pyplot as plt
import cv2
from ultralytics import YOLO

# Load the best.pt
model = YOLO("/content/best.pt")

results = model.predict('/content/car.jpg', device='cpu')


result = results[0]


image_with_boxes = result.plot()  #  (BGR format)

# Convert from BGR to RGB (opencv to matplotlib format)
image_rgb = cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB)

# Display the images

plt.figure(figsize=(10, 8))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Predicted Bounding Box")
plt.show()
```
